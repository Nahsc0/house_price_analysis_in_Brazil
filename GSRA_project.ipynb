```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("brasil-real-estate-1.csv")
```

```python
df1
```

```python
df1.info()
```

```python
df1.dropna()
```

```python
df1.dropna(inplace =True)
```

```python
df1
```

```python
# Split the 'lat-lon' column into two columns 'lat' and 'lon'
df1[['lat', 'lon']] = df1['lat-lon'].str.split(',', expand=True)

# Convert the 'lat' and 'lon' columns to float
df1['lat'] = df1['lat'].astype(float)
df1['lon'] = df1['lon'].astype(float)
```

```python
df1
```

```python
# Extract the state name from the 'place_with_parent_names' column
df1['state'] = df1['place_with_parent_names'].str.extract('\|Brasil\|(.*?)\|')
```

```python
df1
```

```python
# Remove the dollar sign and commas from the 'price_usd' column
df1['price_usd'] = df1['price_usd'].str.replace('$', '')
df1['price_usd'] = df1['price_usd'].str.replace(',', '')

# Convert the 'price_usd' column to float
df1['price_usd'] = df1['price_usd'].astype(float)
```

```python
df1
```

```python
type('price_usd')
```

```python
df1['price_usd']
```

```python
# Drop the 'lat-lon' and 'place_with_parent_names' columns
df1 = df1.drop(columns=['lat-lon', 'place_with_parent_names'])
```

```python
df1
```

```python
# Calculate the average price per square meter for each region and property type
df1_avg_price_per_sqm = df1.groupby(['region', 'property_type'])['price_usd'].mean() / df1.groupby(['region', 'property_type'])['area_m2'].mean()
```

```python
df1_avg_price_per_sqm
```

```python
# Create a pivot table
pivot_table = df1.pivot_table(index='region', columns='property_type', values='price_usd')

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, cmap='coolwarm')
plt.title('Average Price per Square Meter by Region and Property Type')
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df1[['area_m2', 'price_usd']].corr()

# Print the correlation matrix
print(correlation_matrix)
```

```python
# Create a scatter plot
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df1, x='area_m2', y='price_usd')
plt.title('Relationship between Area and Price')
plt.show()
```

```python
# Calculate the average price per square meter for each property type
df1_avg_price_per_sqm = df1.groupby('property_type')['price_usd'].mean() / df1.groupby('property_type')['area_m2'].mean()

colors = ['green', 'blue'] # Add more colors if needed


# Create a bar plot
plt.figure(figsize=(10, 8))
sns.barplot(x=df1_avg_price_per_sqm.index, y=df1_avg_price_per_sqm.values, palette=colors)
plt.title('Average Price per Square Meter by Property Type')
plt.show()
```
